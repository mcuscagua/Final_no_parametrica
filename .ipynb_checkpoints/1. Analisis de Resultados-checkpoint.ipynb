```python
import pandas as pd
import matplotlib.pylab as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
FClasicaL = pd.read_excel('Curvas Clasicas.xlsx')
FClasicaContL = pd.read_excel('Curvas Clasicas Contaminadas.xlsx')
FClasicaLS = pd.read_excel('Curvas Long Short Clasicas.xlsx')
FClasicaContLS = pd.read_excel('Curvas Clasicas Long Short Contaminadas.xlsx')

FRobMargL = pd.read_excel('Curvas Long Robusta Marginal.xlsx')
FRobMargContL = pd.read_excel('Curvas Mediana Marginal Contaminadas.xlsx')
FRobMargLS = pd.read_excel('Curvas Long Short Robusta Marginal.xlsx')
FRobMargContLS = pd.read_excel('Curvas Mediana Marginal Long Short Contaminadas.xlsx')

FRobIndiL = pd.read_excel('Curvas Robusta Indicadora.xlsx')
FRobIndiContL = pd.read_excel('Curvas Mediana Indicadora Contaminadas.xlsx')
FRobIndiLS = pd.read_excel('Curvas Long Short Robusta Indicadora.xlsx')
FRobIndiContLS = pd.read_excel('Curvas Mediana Indicadora Contaminadas.xlsx')
```

```python
plt.figure(figsize= (8, 8))

plt.plot(FClasicaL.iloc[:,1], FClasicaL.iloc[:,0])
plt.plot(FRobMargL.iloc[:,1], FClasicaL.iloc[:,0])
plt.plot(FRobIndiL.iloc[:,1], FClasicaL.iloc[:,0])

plt.legend(['media - matriz de varianzas y covarianzas',
            'mediana marginal - matriz de Ledoit Wolf',
            'mediana indicadora - matriz de Ledoit Wolf'])

plt.grid()
plt.show()
```
